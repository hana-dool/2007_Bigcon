```python
# dataset train/test set 으로 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
def MAPE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

# LGB

```python
import lightgbm as lgb
train_ds = lgb.Dataset(X_train, label = y_train)
params = {'learning_rate': 0.01,
          'max_depth': 16,
          'boosting': 'gbdt',
          'objective': 'regression',
          'metric': 'mse',
          'is_training_metric': True,
          'num_leaves': 144,
          'feature_fraction': 0.9,
          'bagging_fraction': 0.7,
          'bagging_freq': 5,
          'seed':2020}

model = lgb.train(params, train_ds, 1500, verbose_eval=100)
```

```python
y_pred=model.predict(X_test)
```

```python
MAPE(y_pred,y_test)
```

```python
from sklearn.model_selection import train_test_split
X_train_, X_valid_, y_train_, y_valid_ = train_test_split(X_train, y_train, test_size=0.20, random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation,BatchNormalization
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras import metrics
```

```python
X_train.shape
```

```python
model_1 = keras.models.Sequential()
model_1 = Sequential([
    Dense(128, kernel_initializer='normal', activation = "relu", input_shape=X_train.shape[1:]), 
    Dense(256, kernel_initializer='normal', activation = "relu"),
    Dense(512, kernel_initializer='normal', activation = "relu"),
    Dense(1024, kernel_initializer='normal', activation = "relu"),
    Dense(512, kernel_initializer='normal', activation = "relu"),
    Dense(256, kernel_initializer='normal', activation = "relu"),
    Dense(32, kernel_initializer='normal', activation = "relu"),
    Dense(1, kernel_initializer='normal'), 
])

model_1.compile(loss="mape",  # 평가기준이 mape 이니까 이걸로 하자.
              optimizer=keras.optimizers.Adam())
history_1 = model_1.fit(X_train_, y_train_, epochs = 20,validation_data=(X_valid_, y_valid_))
```

```python
evaluation_1
```

```python
model_2 = keras.models.Sequential()
model_2 = Sequential([
    Dense(128, activation='relu',input_shape=X_train.shape[1:]),
    Dense(256 , activation = "relu"),
    Dense(256 , activation = "relu"),
    Dense(128 , activation = "relu"),
    Dense(32 , activation = "relu"),
    Dense(1)
])

model_2.compile(loss="mape",  # 평가기준이 mape 이니까 이걸로 하자.
              optimizer=keras.optimizers.Adam())
history_2 = model_2.fit(X_train_, y_train_,batch_size=64, epochs=40, validation_data=(X_valid_, y_valid_))
evaluation_2 = model_2.evaluate(X_test, y_test)
```

```python
# deeper model
model3 = keras.models.Sequential()
model3.add(keras.layers.Dense(200, activation="selu",kernel_initializer="lecun_normal",input_shape=X_train.shape[1:]))
for layer in range(5):
    model3.add(keras.layers.Dense(200, activation="selu",kernel_initializer="lecun_normal"))
for layer in range(5):
    model3.add(keras.layers.Dense(300, activation="selu",kernel_initializer="lecun_normal"))
for layer in range(5):
    model3.add(keras.layers.Dense(100, activation="selu",kernel_initializer="lecun_normal"))
model3.add(keras.layers.Dense(10))
model3.add(keras.layers.Dense(1))
model3.compile(loss="mape",  # 평가기준이 mape 이니까 이걸로 하자.
              optimizer=keras.optimizers.Adam())
history3 = model3.fit(X_train_, y_train_,batch_size=64 ,epochs=40, validation_data=(X_valid_, y_valid_))
evaluation3 = model3.evaluate(X_test, y_test)
```

```python
model4 = keras.models.Sequential()
model4.add(keras.layers.Dense(100, activation="relu",kernel_initializer="he_normal",input_shape=X_train.shape[1:]))
for layer in range(10):
    model4.add(keras.layers.Dense(100, activation="relu",kernel_initializer="he_normal"))
for layer in range(10):
    model4.add(keras.layers.Dense(300, activation="relu",kernel_initializer="he_normal"))
for layer in range(10):
    model4.add(keras.layers.Dense(100, activation="relu",kernel_initializer="he_normal"))
model4.add(Dropout(0.5)) # overfitting 방지
model4.add(keras.layers.Dense(10))
model4.add(keras.layers.Dense(1))
model4.compile(loss="mape",  # 평가기준이 mape 이니까 이걸로 하자.
              optimizer=keras.optimizers.Adam())
history4 = model4.fit(X_train_, y_train_, epochs=40, validation_data=(X_valid_, y_valid_))
evaluation4 = model4.evaluate(X_test, y_test)
```

```python
model5 = keras.models.Sequential()
model5.add(keras.layers.Dense(50, activation="selu",kernel_initializer="lecun_normal",input_shape=(36,)))
for layer in range(3):
    model5.add(keras.layers.Dense(100, activation="selu",kernel_initializer="lecun_normal"))
keras.layers.BatchNormalization()
for layer in range(3):
    model5.add(keras.layers.Dense(300, activation="selu",kernel_initializer="lecun_normal"))
keras.layers.BatchNormalization()
for layer in range(3):
    model5.add(keras.layers.Dense(500, activation="selu",kernel_initializer="lecun_normal"))
keras.layers.BatchNormalization()
for layer in range(3):
    model5.add(keras.layers.Dense(300, activation="selu",kernel_initializer="lecun_normal"))
keras.layers.BatchNormalization()
for layer in range(3):
    model5.add(keras.layers.Dense(100, activation="selu",kernel_initializer="lecun_normal"))
keras.layers.BatchNormalization(),
model5.add(keras.layers.Dense(30))
model5.add(keras.layers.Dense(10))
model5.add(keras.layers.Dense(1))
model5.compile(loss="mape",  # 평가기준이 mape 이니까 이걸로 하자.
              optimizer=keras.optimizers.Adam())
history5 = model5.fit(X_train_, y_train_, epochs=20, validation_data=(X_valid_, y_valid_))
evaluation5 = model5.evaluate(X_test, y_test)
```
